```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def solve_quadratic_equation(a, b, c):
    delta = b**2 - 4*a*c

    if delta > 0:
        x1 = (-b + np.sqrt(delta)) / (2*a)
        x2 = (-b - np.sqrt(delta)) / (2*a)
        print("Solução 1:", x1)
        print("Solução 2:", x2)
        plot_solutions(a, b, c, x1, x2)
    elif delta == 0:
        x = -b / (2*a)
        print("Solução única:", x)
        plot_solutions(a, b, c, x)
    else:
        print("Não há solução real.")

def plot_solutions(a, b, c, *solutions):
    x = np.linspace(-10, 10, 400)
    y = a*x**2 + b*x + c

    plt.plot(x, y, label=f"{a}x^2 + {b}x + {c} = 0")
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True, linestyle='--', linewidth=0.5)

    for solution in solutions:
        plt.plot(solution, 0, 'ro')

    plt.legend()
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Equação do Segundo Grau')
    plt.show()


a = float(input("Digite o coeficiente 'a': "))
b = float(input("Digite o coeficiente 'b': "))
c = float(input("Digite o coeficiente 'c': "))
print("-------------------------------")


solve_quadratic_equation(a, b, c)
```
